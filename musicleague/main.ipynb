```python
def get_playlist_tracks(playlist_id):
    """
    Fetches tracks from a Spotify playlist.
    Args:
        playlist_id (str): The Spotify playlist ID.
    Returns:
        list of tuples: List containing tuples of (song name, artist name).
    """
    tracks = []
    results = sp.playlist_items(playlist_id, additional_types=['track'])
    while results:
        for item in results['items']:
            track = item['track']
            if track:
                song_name = track['name']
                artist_name = ", ".join(artist['name'] for artist in track['artists'])
                tracks.append((song_name, artist_name))
        results = sp.next(results) if results['next'] else None
    return tracks

def save_to_csv(data, filename="playlist_songs.csv"):
    """
    Saves the list of songs and artists to a CSV file.
    Args:
        data (list of tuples): List containing tuples of (song name, artist name).
        filename (str): The name of the output CSV file.
    """
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Song Name", "Artist Name"])  # Header
        writer.writerows(data)


playlist_ids = [
    "5B6gCfDNSRsIXPF07cy0Dd",
    "37TgBAsDXGObTHqaNQi4bz",
    "4AZL34bwUEL6b0RikQjQRJ",
    "2IF0QEItDwDm5CJjCvFIiz",
    "5aF5a6PgFY0W9v7UXniOXg",
    "4nEKcDIj5o1onWM2wUNW62",
    "0dP6vIZSE4cAICru0NfojJ",
    "5srS9GBlVKRxz3jES0vJoQ",
    "4wfB2Pnhnahd9RoKGEujVv"]

all_tracks = []
for playlist_id in playlist_ids:
    print(f"Fetching tracks from playlist {playlist_id}...")
    all_tracks.extend(get_playlist_tracks(playlist_id))

save_to_csv(all_tracks)
print("Tracks saved to 'playlist_songs.csv'.")
```
